```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
dataset=load_breast_cancer()
```

```python
dataset
```

```python
dataset.keys()
```

```python
dfcancer=pd.DataFrame(dataset.data,columns=dataset.feature_names)
dfcancer
```

```python
dfcancer['target']=pd.DataFrame(dataset.target)
dfcancer
```

```python
x=dfcancer.iloc[:,0:-1]
y=dfcancer.iloc[:,-1]
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.20,random_state=45)
```

```python
lg=LogisticRegression()
```

```python
lg.fit(train_x,train_y)
```

```python
lg.score(train_x,train_y)
#How well the train_x,train_y data has fit into the dataset
```

```python
pred=lg.predict(test_x)
```

```python
print(accuracy_score(test_y,pred))
```

```python
print(confusion_matrix(test_y,pred))
```

```python
print(classification_report(test_y,pred))
```

```python
print(accuracy_score(test_y,pred))
print(confusion_matrix(test_y,pred))
print(classification_report(test_y,pred))
```
