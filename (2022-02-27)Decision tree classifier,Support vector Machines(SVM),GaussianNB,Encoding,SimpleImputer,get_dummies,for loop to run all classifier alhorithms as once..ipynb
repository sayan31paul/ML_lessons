# Decision Tree Classifier

The leaf nodes are the target or the level.

```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
iris=load_iris()
iris.keys()
```

```python
ds=pd.DataFrame(data=iris.data)
ds
```

```python
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.22,random_state=43)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#DecisionTreeClassifier(criterion='gini').............-->Default
#DecisionTreeClassifier(criterion='entropy')
dtc=DecisionTreeClassifier() #creating the object/instance of decision tree.
dtc.fit(x_train,y_train)
dtc.score(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

# Simple Imputer

```python
from sklearn.impute import SimpleImputer

imp=SimpleImputer(strategy='most_frequent')   #creating new object from SimpleImputer class with required strategy.

ds['City']=imp.fit_transform(ds['City'].values.reshape(-1,1)) 
ds

#    .values.reshape(-1,1) converts 1D array to 2D array.
#  we can use strategy='most_frequent' only in case of string data
```

Here the NaN values in  'City' column  are replaced with  'Hongkong' which is the most frequent value in the column.
