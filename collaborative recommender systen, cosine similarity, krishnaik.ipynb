```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
books=pd.read_csv('BX-Books.csv',sep=';',error_bad_lines=False,encoding='latin-1')
users=pd.read_csv('BX-Users.csv',sep=';',error_bad_lines=False,encoding='latin-1')
ratings=pd.read_csv('BX-Book-Ratings.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
books.columns=['ISBN', 'bookTitle', 'bookAuthor', 'yearOfPublication', 'publisher',
       'imageUrlS', 'imageUrlM', 'imageUrlL']

#renaming the columns
```

```python
users.columns=['userID', 'Location', 'Age']

#renaming the columns
```

```python
ratings.columns=['userID', 'ISBN', 'bookRating']

#renaming the columns
```

```python
books=books[['ISBN', 'bookTitle']]
books

#books contains the ISBN and the book name
```

```python
users
```

```python
ratings
```

```python
ratings=pd.merge(ratings,books,on='ISBN')
ratings
```

### To ensure statistical significance,users with less than 200 ratings and books with less than 100 ratings are excluded

```python
counts1=ratings['userID'].value_counts()
counts1

#How many times each userID has appeared in the dataset.
```

```python
ratings=ratings[ratings['userID'].isin(counts1[counts1>=200].index)]
ratings
```

```python
counts=ratings['ISBN'].value_counts()
counts

#Which book has got how many ratings
```

```python
ratings=ratings[ratings['ISBN'].isin(counts[counts>=100].index)]
ratings
```

```python
ratings=ratings.dropna(axis=0,subset=['bookTitle'])
ratings

#Drop the rows if the bookTitle is NaN
#dropna drops rows or columns with NaN values
#drop() drops selected rows or columns
#Luckily none of the book titles were NaN. So no rows were dropped
```

```python
user_rating_pivot=ratings.pivot(index='bookTitle',columns='userID',values='bookRating').fillna(0)
user_rating_pivot
#.fillna(0) fills the NaN values with zero.
```

The above pivot table is very sparse. Most of the values are zero. It is better to convert it into sparse matrix for the ease of computation

```python
from scipy.sparse import csr_matrix
user_rating_matrix=csr_matrix(user_rating_pivot.values)
```

```python
user_rating_matrix
```

```python
from sklearn.neighbors import NearestNeighbors

model_knn=NearestNeighbors(metric='cosine',algorithm='brute') #object creation from class
# algorithm='brute' means check every row with every row.
model_knn.fit(user_rating_matrix) #not entering any train_y value. Hence unsupervised ML
```

```python
query_index=np.random.choice(user_rating_pivot.shape[0]) #choose any random index from the pivot table
query_index
```

```python
distances,indices=model_knn.kneighbors(user_rating_pivot.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)

#values.reshape(1,-1) converts 1D to 2D
#values.reshape(1,-1) converts Series to DataFrame
```

```python
for i in range(0,len(distances.flatten())):
    if i==0:
        print('Recommendations for',user_rating_pivot.index[query_index],'is :')
    else:
        print(i,user_rating_pivot.index[i],distances.flatten()[i])
```
