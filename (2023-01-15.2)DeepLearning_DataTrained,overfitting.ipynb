```python
"""As the number of layers increases, the number of trainable parameters increases which results in overfitting.
Some techniques to avoid overfitting are:
1. Batch normalisation 
2. Dropout 
3. Weight initialisers or kernel initialisers
4. Early stopping
"""
```

```python
"""Batch normalisation:
In the training data, one image might be very diferent from the other image. Ex- our model is totally trained on white dogs but we get a black dog. The 
distribution for white dog will be different from black dog. In that case we require something where despite the difference in distribution our model 
is able to work. Batch normalisation helps to normalise and keep them within the same distribution.
Batch normalisation aslo helps in converging data. So we can get to the results faster."""
```

```python
"""Dropout:
Let us have a drop out value of 0.2. Randomly 20% of the neurons will be removed from each layer. This is known as thinning out of layers."""
```
