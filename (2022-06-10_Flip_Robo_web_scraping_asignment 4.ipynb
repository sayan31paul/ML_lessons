```python
dataset_name=[]
data_type=[]
task=[]
attribute_type=[]
no_of_instances=[]
no_of_attribute=[]
year_note=[]
for i in range(2,624):
    try:
        name=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[1]/table/tbody/tr/td[2]/p/b/a')
        dataset_name.append(name.text)
    except NoSuchElementException:
        dataset_name.append('--')
    
    try:
        element=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[2]/p')
        data_type.append(element.text)
    except NoSuchElementException:
        data_type.append('--')

    try:
        element=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[3]/p')
        task.append(element.text)
    except NoSuchElementException:
        task.append('--')
    
    try:
        element=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[4]/p')
        attribute_type.append(element.text)
    except NoSuchElementException:
        attribute_type.append('--')
    
    try:
        element=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[5]/p')
        no_of_instances.append(element.text)
    except NoSuchElementException:
        no_of_instances.append('--')
        
    try:
        element=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[6]/p')
        no_of_attribute.append(element.text)
    except NoSuchElementException:
        no_of_attribute.append('--')
        
    try:
        element=driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i)+']/td[7]/p')
        year_note.append(element.text)
    except NoSuchElementException:
        year_note.append('--')
```
