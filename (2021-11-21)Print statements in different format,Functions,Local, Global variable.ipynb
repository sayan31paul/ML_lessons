# Functions with parameters

```python
def additionfunc(a,b):
    c=a+b
    print(c)
```

```python
additionfunc()
```

```python
additionfunc(10,20)
```

```python
additionfunc(50,85)
```

```python
additionfunc("Data","Science")
```

```python
additionfunc("Data ",'Science')
```

```python
#For loop can also run inside a function
def runfunc():
    print('inside function')
    for i in [1,2,3,4,5]:
        print(i)
```

```python
runfunc()
```

```python
#A user defined function can also be called inside an user defined function
def outside_func():
    print('Outside function')
    runfunc()                  #calling runfunc inside outside_func
```

```python
outside_func()
```

```python
def additionfunc(a,b):
    c=a+b
    print(c)
```

```python
#User input values being passed into function values
def user_add():
    a=int(input('enter first number'))
    b=int(input('enter second number'))
    additionfunc(a,b)
```

```python
user_add()
```
