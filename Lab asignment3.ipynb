```python
def check_posn(x,y):
    #intersection of line 2 and line 3 is 10/11
    #intersection of line 1 and line 4 is -8.6
    
    z=[] #contains the vale obtained by putting x and y in the equations.
    z.append(2.4-y)
    z.append(3*x-y+5)
    z.append(-2.5*x-y+10)
    z.append(x-y+11)
    pos_count=0
    neg_count=0
    for i in range(4):
        if z[i]>0:
            pos_count +=1
        elif z[i]<0:
            neg_count +=1
        else:
            pos_count +=1
            
    if pos_count==neg_count:
        if x<(10/11) and x>-8.6 :
            print("lie between the given lines")
        else:
            print("doesnot lie between the given lines")
            
    else:
        print("Does not lie between the given lines")
    
```

```python
check_posn(-1,3)
```

```python
check_posn(7.5,5)
```

```python
check_posn(10,20)
```

```python
fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d') #object creation

x = np.linspace(-10,10,20)
y = np.linspace(-10,10,20)

X, Y = np.meshgrid(x, y)
Z = X+Y+X*Y+3

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)

# Set axes label
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
```
