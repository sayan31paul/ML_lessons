```python
"""Convolutional neural networks"""
"""CNN is basically used for any image related tasks. In MLP we had to flatten an image so that it can fit into the MLP. In case we have to use both dimensions
of an image we have to use convolution neural networks. CNN is used for tasks like image classification, object detection."""
```

```python
"""In mnist dataset we had black and white image so he had only one channel. But in coloured images we have three different channels i.e RGB resepctively.
The thickness of the image represents the RGB channel. The thickness is present only in coloured images."""
```

```python
"""Edge detection :
It detects all the edges inside the image. Higher is the contrast in the image, more easily we can detect the edges inside the image. If edge is horozontal we
will have to use a horizontal edge detector, for vertical egdes use vertical edge detector. The edge detector is also known as kernel or filter.
Finding edges from a picture is one type of convolution.
Edges of picture can also be called as feature of a picture.
Egde detector helps us find a feature of a picture.
Kernels are used to find the feature of an image."""
```

These are the sobel edge detectors, they are used to detect horizontal and vertical lines.

```python
"""Introduction to convolution"""
"""Features of an image include edges, corners.
Changing values inside the edge detector can help us identify other features of an image.
Values inside the edge detector/kernel/filters are the trainable parameters in CNN. They are similar to weights and biases in MLP.

What are the features that we can learn in a dog? -> pointy ears, tongue sticking out , two eyes. """
```

```python
"""Coloured images.
Coloured images has a dimension of l*b*3 where l in the number of pixels in length, b is the number of pixels in breadth and 3 represents the three channles
R, G, B resepctively."""
```

```python
"""The result depends on three parameters i.e size, stride and distance.
Size is the 2D dimension of the image.
Stride the number of pixels by which we are moving the edge_detector/kernel/filter.
Padding is the white pixel added along the perimeter of the image. Padding is applied to avoid loss of information"""

"""Padding is of two types:-
a.Valid b.Same

Valid padding:- No padding is known as valid padding
Same padding:- It gets padded as many times as required such that the input image dimension is same as the result dimension.

Padding is done to avoid the loss of border valus during convolution."""
```

```python
"""Cloured images will have some depth because of the RGB channels. Because of the depth, the filter/kernel/egde_detector will also have some depth.
i.e we are going to have 3 dimensional kernel. Now the filter will also be having three channels i.e red,green,blue.
Red channel of the image will convolve with the red channel of the kernel and so on.
We will still be getting a two dimensional result. We will have to take the sum of all the convolutions and add them and put them in the result."""
```
