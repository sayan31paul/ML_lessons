```python
import ast              #converts string indices to int
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):          #ast.literal_eval(obj) converts strings to integers.
        L.append(i['name'])
    return L
```

```python
#We have to take the first 3 cast from each row of 'cast' column
import ast                    #converts string indices to int
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):  #ast.literal_eval converts string indices to int
        if counter!=3:
            L.append(i['name'])
            counter=counter+1
        else:
            break
    return L
```

```python
import ast
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=="Director":
            L.append(i['name'])
            break
    return L
       
            
        
```

# Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movies['tags'])
vectors
```

```python
vectors.shape
```

```python
#Lets see what are the top 5000 words
cv.get_feature_names()
```

```python
#Certain words are present in the format ['love','loved','loves']
#We have to convert it to ['love','love','love']
#This will help to include some more words in the top 5000 words.
#This will be done by PorterStemmer() class
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
```

```python
ps.stem('loving')
```

```python
ps.stem('loved')
```

```python
def stem(text):    #text is the argument the function is going to take.
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

#.split() breaks a string into multiple strings and stored them in alist
#.join(y) joins the strings after applying PorterStemmer
        
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
vectors=cv.fit_transform(new_movies['tags'])
vectors
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=new_movies[new_movies['title']==movie].index[0]
    movies_list=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        
        print(i[0])
```

```python
recommend('Avatar')
```
