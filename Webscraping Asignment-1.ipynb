# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://www.dineout.co.in/') 
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
titles=[]     #empty list for storing the titles

for i in soup.find_all('div',class_="_9FNkO"):
    titles.append(i.text)
```

```python
titles
```

```python
location=[]     #empty list for storing the titles

for i in soup.find_all('div',class_="_9FNkO"):
    location.append(i.text.replace('Connaught',' Connaught_Place').split()[-4:])
for i in location:
    i.pop(1)
```

```python
location
```

```python
rating=[] #empty list

for i in soup.find_all('div',class_='kGUdK _1oTbl'):
    rating.append(i.text)
```

```python
rating
```

```python
images=[]
for i in soup.find_all('img',class_='_3ZUwv'):
    images.append(i['data-src'])
    
```

```python
images
```

```python
import pandas as pd
df=pd.DataFrame({'Titles':titles,'Location':location,'Rating':rating})
```
