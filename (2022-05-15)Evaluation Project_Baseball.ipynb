# Removing outliers

# Training process begins

```python
x=df.iloc[:,0:-1]
```

```python
x.head(2)
```

```python
y=df.iloc[:,-1]
```

```python
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import GridSearchCV
```

Lets try out the best polynomial for LinearRegression

```python
max_cv_score=0
best_polynomial=0
for i in range(1,4):
    poly = PolynomialFeatures(i)
    X_=poly.fit_transform(x)
    lr =LinearRegression()
    lsscore=cross_val_score(lr,X_,y,cv=5)
    if lsscore.mean()>max_cv_score:
        max_cv_score=lsscore.mean()
        best_polynomial=i

print(max_cv_score)
print(best_polynomial)
```

Finding out the random_state for which LinearRegression model works best

```python
max_r2_score=0
best_rand_state=0
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    
    if r2_score(y_test,y_pred)>max_r2_score:
        max_r2_score=r2_score(y_test,y_pred)
        best_rand_state=i

print(max_r2_score)
print(best_rand_state)
```

Training the Linear Regression model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=83)
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print("r2_score is: ",r2_score(y_test,y_pred))
```

Lets find out the best parameters for ridge regression model

```python
r=Ridge()
parameters={'alpha':[0.1,0.01,0.001],'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs']}
gd_sr=GridSearchCV(estimator=r, param_grid= parameters, scoring='r2')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

Trying out the best random_state for RidgeRegression model for the above obtained parameters.

```python
max_r2_score=0
best_rand_state=0
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    r=Ridge(alpha= 0.001, solver= 'sparse_cg')
    r.fit(x_train,y_train)
    y_pred=r.predict(x_test)
    
    if r2_score(y_test,y_pred)>max_r2_score:
        max_r2_score=r2_score(y_test,y_pred)
        best_rand_state=i

print(max_r2_score)
print(best_rand_state)
```

Lets train the final Ridge regression model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=83)
r=Ridge(alpha= 0.001, solver= 'sparse_cg')
r.fit(x_train,y_train)
y_pred=r.predict(x_test)
print("r2_score is: ",r2_score(y_test,y_pred))


#Ridge Regresion model is giving slightly higher accuracy than LinearRegression model
```

Finding out the best parameters for Lasso Regression model

```python
l=Lasso()
parameters={'alpha':[0.1,0.01,0.001]}
gd_sr=GridSearchCV(estimator=l, param_grid= parameters, scoring='r2')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

Finding out the best random_state for LassoRegression

```python
max_r2_score=0
best_rand_state=0
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    l=Lasso(alpha=0.1)
    l.fit(x_train,y_train)
    y_pred=l.predict(x_test)
    
    if r2_score(y_test,y_pred)>max_r2_score:
        max_r2_score=r2_score(y_test,y_pred)
        best_rand_state=i

print(max_r2_score)
print(best_rand_state)
```

Training the final model for LassoRegression

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=83)
l=Lasso(alpha=0.1)
l.fit(x_train,y_train)
y_pred=l.predict(x_test)
print("r2_score is: ",r2_score(y_test,y_pred))


#Lasso Regression is giving better results than Linear and Ridge Regresison
```

Applying GridSaerchCV on KNN Regressor

```python
knr=KNeighborsRegressor()
parameters={'n_neighbors':[2,3,4,5,7],"weights":['uniform','distance'],'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],}
gd_sr=GridSearchCV(estimator=knr, param_grid= parameters,scoring='r2')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

finding out the best random state for KNeighborsregressor

```python
max_r2_score=0
best_rand_state=0
for i in range(0,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    knr=KNeighborsRegressor(algorithm= 'auto', n_neighbors= 5, weights= 'uniform')
    knr.fit(x_train,y_train)
    y_pred=knr.predict(x_test)
    
    if r2_score(y_test,y_pred)>max_r2_score:
        max_r2_score=r2_score(y_test,y_pred)
        best_rand_state=i

print(max_r2_score)
print(best_rand_state)
```

r2_score obtained from KNeighborsRegressor in less as comapred to the Linear,Ridge, Lasso Regression models. So we are not training any model from KNeighborsRegressor

Finding out the best parameters for RandomForestRegressor

```python
rfr=RandomForestRegressor()
parameters={'n_estimators':[20,30,50,70],'max_depth':[10,13,15,20]}
gd_sr=GridSearchCV(estimator=rfr, param_grid= parameters,scoring='r2')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

RandomForestRegressor is giving very bad cross_validation_score. So we are not approaching further with RandomForestRegressor.

# Concluding the final model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=83)
l=Lasso(alpha=0.1)
l.fit(x_train,y_train)
y_pred=l.predict(x_test)
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2_score is: ",r2_score(y_test,y_pred))
```
