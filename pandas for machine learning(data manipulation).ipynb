# How to rename categories using pandas method

```python
new_survived=pd.Categorical(titanic_train["Survived"]) #categories available in the column will be stored in new_survived
new_survived=new_survived.rename_categories(["Died","Survived"])
new_survived.describe()
```

```python
pd.Categorical(titanic_train["Survived"])
```

```python
titanic_train
```

```python
#trying renaming categories with another column
titanic_train["Pclass"].unique()
```

```python
new_Pclass=pd.Categorical(titanic_train["Pclass"],ordered=True) #arranges the classes in ascending order so thet the new names are asigned correctly
new_Pclass
```

```python
new_Pclass=new_Pclass.rename_categories(['class1','class2','class3'])
new_Pclass.describe()
```

```python
titanic_train #It is evident that new category names have not been asigned in the respective columns.
```

```python
#reasign new category names to the respective columns.
titanic_train["Pclass"]=new_Pclass
```

```python
titanic_train["Survived"]=new_survived
```

```python
titanic_train #changes in category names is completed successfully
```

```python
#trying renaming categories with another column
titanic_train["Cabin"].unique()
```

```python
char_cabin=titanic_train["Cabin"].astype(str) #convert data to str
#List comprehension( Creating list using another set of list, string etc)
new_Cabin=np.array([cabin[0] for cabin in char_cabin]) #Take first letter
new_Cabin=pd.Categorical(new_Cabin)
new_Cabin.describe()
```

# Find the maximum fare

```python
index=np.where(titanic_train["Fare"]==max(titanic_train["Fare"])) #np.where give the index number.
index
```

```python
titanic_train.loc[index] #loc() takes index number as parameter
```

# Find the family who has the highest members

```python
#Sibsp and Parch are siblings and parents/gurdains respectively(add both of them and create new column)
titanic_train["Family"]=titanic_train['SibSp']+titanic_train['Parch'] #adding of columns
most_family=np.where(titanic_train["Family"]==max(titanic_train["Family"])) #np.where gives the index numbers.
titanic_train.loc[most_family] #loc() takes index numbers as input
```
