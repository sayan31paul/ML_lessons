# Operating on null values

```python
#Create dataframe with missing values
raw_data={'first_name':['Jason',np.nan,'Tina','Jake','Any','Madhu'],
         'last_name':['Miller',np.nan,'Aki','Milner','Cooze','Patil'],
         'age':[42,np.nan,36,24,73,25],
         'sex':['M',np.nan,'F','M','F','F'],
         'unit-1':[4,np.nan,np.nan,2,3,4],
         'unit-2':[3,np.nan,np.nan,4,4,4]}
df=pd.DataFrame(raw_data)
df
```

```python
#Drop rows with NaN values
df_no_missing=df.dropna() #This doesnot create any change in the original dataset
df_no_missing
```

```python
#Drop columns with NaN values
df_no_missing=df.dropna(axis=1) #This doesnot create any change in the original dataset
df_no_missing
```

```python
#only rows with all NaN values will be removed
df_cleaned=df.dropna(how='all')
df_cleaned
```

```python
#create a new column filled with missing values
df['Final_Score']=np.nan
df
```

```python
#Drop coulmn if they contain only missing values
df.dropna(axis=1,how='all')
```

```python
#fill all NaN values with 0
df.fillna(0)
df
```

```python
df.fillna(0)
```

```python
df   #Change due to fillna() is not permanent. #Final score column is not permanently deleted instead of dropna()
```

```python
# Deleting the column with NaN value permanently
df.dropna(axis=1,how='all',inplace=True)#inplace=True means that column will be deleted permanently
```

```python
df #Column with all NaN values is deleted permanently.
```

```python
#create a new column filled with missing values
df['Final_Score']=np.nan
df
```

```python
#Fill in NaN values of unit-1 with the mean value of unit-1
df["unit-1"].fillna(df["unit-1"].mean(),inplace=True)
```

```python
df
```

```python
#Fill in missing in unit-2 with each sex's mean value of unit-2 score
df["unit-2"].fillna(df.groupby("sex")["unit-2"].transform("mean"),inplace=True)
```

```python
df
```

```python
#Select some rows but ignore the missing data points
```

```python
df[df['age'].notnull()& df['sex'].notnull()] #effect not permanent. use inplace=True to make it permanent
```

```python
#back-fill
df.fillna(method='bfill') #effect not permanent. use inplace=True to make it permanent
```

```python
#forward-fill
df.fillna(method='ffill') #effect not permanent. use inplace=True to make it permanent
df
```

```python
#making the effect of forward fill permanent
df.fillna(method='ffill',inplace=True)
df
```

```python
# Add values in unit-1 and unit-2 to replace NaN in Final_Score column
df["Final_Score"].fillna(df["unit-1"]+df["unit-2"],inplace=True)
```

```python
df
```

```python
#adding a new column with data
df['Location']='Bengaluru'
```

```python
df
```

# Hierarchical indexing

```python
raw_data={'Company':['Google','Google','Google','Google','Facebook','Facebook','Facebook','Facebook','Amazon','Amazon','Amazon','Amazon'],
         'Project':['1st','1st','2nd','2nd','1st','1st','2nd','2nd','1st','1st','2nd','2nd'],
         'Incharge':['Sunder','Ruth','Benjamin','Sergey','Mark','Shery','Eduardo','David','Jeffrey','Andy','Keith','Edith'],
         'Sales(in millions)':[54,24,31,12,23,24,34,31,32,23,52,13],
         'Profit(in millions)':[14,8,9,4,10,11,14,13,12,7,22,4]}
df=pd.DataFrame(raw_data)
df
```

```python
#set hierachical index
df.set_index(['Company','Project']) #effect is not permanent
```

```python
#Set the hierarchical index but leave the columns in place
df.set_index(['Company','Project'],drop=False)  #Effect is not permanent
```

```python
df.set_index(['Company','Project'],inplace=True) # Effect is permanent
```

```python
df #Showing the permanent effect
```

```python
#swapping the indexes
df.swaplevel('Company','Project') #use inplace=True to make the effect permanent
```

```python
#summarize the results by company
df.sum(level='Company')
```

```python
#summarize the results by project
df.sum(level='Project')
```

```python
#summarize the results by company and project
df.sum(level=['Company','Project'])
```
