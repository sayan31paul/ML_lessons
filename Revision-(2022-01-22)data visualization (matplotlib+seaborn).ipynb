```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
a=np.array(range(1,8))
b=np.array(range(1,8))
```

```python
print(a)
print(b)
```

```python
plt.title("linear line")
plt.xlabel("a")
plt.ylabel("b")
plt.plot(a,b)      #This doing the actual plotting
plt.show()
```

```python
g=np.array([3,6,2,87,56])
h=np.array([22,5,56,82,55])
plt.xlabel('g')
plt.ylabel('h')
plt.plot(g,h)      #This is doing the actual plotting
plt.show
```

```python
#plot the above plot in broken line in green colour
plt.plot(g,h,'--g')
```

```python
#plotting a bar graph
x=[5,8,10,2,6,12]
y=[12,16,6,10,14,6]
plt.bar(x,y)
plt.title('Bar Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#plotting the above bar graph with green color and alignment should be center
x=[5,8,10,2,6,12]
y=[12,16,6,10,14,6]
plt.bar(x,y,color='g',align='center')
plt.title('Bar Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#aligning the above graph on one side
x=[5,8,10,2,6,12]
y=[12,16,6,10,14,6]
plt.bar(x,y,color='g',align='edge')          #align on the side.
plt.title('Bar Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,2,79,31,27])
plt.hist(a,bins=[0,20,40,60,80,100])       #doing the actual plotting
plt.title('histogram')
plt.show()
```

# IRIS DATASET

```python
dsiris=pd.read_csv('iris.csv')
dsiris
```

```python
sns.set(style='whitegrid') #just for a visually attractive presentation
ax=sns.stripplot(x='class',y='sepal length',data=dsiris) #plotting actually
plt.title('Graph')
plt.show
```

```python
ax=sns.swarmplot(x='class',y='sepal length',data=dsiris) #plotting actually
plt.title('Graph')
plt.show
```

```python
ax=sns.swarmplot(x='class',y='petal length',data=dsiris) #plotting actually
plt.title('Graph')
plt.show
```
