# To ensure statistical significance users with less than 200 ratings and books with less than 100 ratings are excluded

```python
counts1=ratings['userID'].value_counts()
ratings=ratings[ratings['userID'].isin(counts1[counts1>=200].index)]
counts=ratings['ISBN'].value_counts()
ratings=ratings[ratings['ISBN'].isin(counts[counts>=100].index)]
```

# Collaborative filtering using K-Nearest Neighbors(KNN)

```python
combine_book_rating=pd.merge(ratings,books,on='ISBN')
columns=['yearOfPublication','publisher','bookAuthor','imageurlS','imageurlM','imageurlL']
combine_book_rating=combine_book_rating.drop(columns,axis=1)
combine_book_rating
```

```python
combine_book_rating=combine_book_rating.dropna(axis=0,subset=['bookTitle'])
combine_book_rating

#Drop the rows if the bookTitle is NaN in that row.
#dropna drop rows or columns with NaN values
```

We then groupby booktiles and create a new column for total rating count

```python
book_ratingCount = (combine_book_rating.
     groupby(by = ['bookTitle'])['bookRating'].
     count().
     reset_index().
     rename(columns = {'bookRating': 'totalRatingCount'})
     [['bookTitle', 'totalRatingCount']]
    )

book_ratingCount
```

```python
rating_with_totalRatingCount=pd.merge(combine_book_rating,book_ratingCount,on='bookTitle')
rating_with_totalRatingCount
```

# Implementing KNN

We convert our table to a 2D matrix, and fill the missing values with zeros(since we have to find the dot product between vectors and also we have to find the modulus of the vectors). We then tranform the values(ratings) of the matrix dataframe into scipy sparse matrix for more efficient calculations.

Finding the nearest neighbors we use unsuprevised algorithms with sklearn neighbors. The algorithm we use to compute the nearest neighbors is "brute" and we specify 'metric='cosine' so that the algorithm will calculate the cosine similarity netween rating vectors. Finally, we fit the model.

```python
from scipy.sparse import csr_matrix
user_rating_pivot=rating_with_totalRatingCount.pivot(index='bookTitle',columns='userID',values='bookRating').fillna(0)
user_rating_pivot
```

```python
user_rating_matrix=csr_matrix(user_rating_pivot.values)
user_rating_matrix
```

```python
from sklearn.neighbors import NearestNeighbors

model_knn=NearestNeighbors(metric='cosine',algorithm='brute') #creating object from class
model_knn.fit(user_rating_matrix) #not entering any train_y, hence this is unsupervised ML.
```

```python
#The ML model has already been created.
#Now we will be selecting one random movie

query_index=np.random.choice(user_rating_pivot.shape[0])
query_index
```

```python
distances,indices=model_knn.kneighbors(user_rating_pivot.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)
```

```python
for i in range(0,len(distances.flatten())):
    if i==0:
        print('Recommendations for',user_rating_pivot.index[query_index],'is :')
    else:
        print(i,user_rating_pivot.index[i],distances.flatten()[i])
        
# .flatten() can flatten a matrix to one dimension in python
```
