```python
#Putting the artist layer to use
#Lets try to generate a histogram using the artist layer.
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas   #import FigureCanvas
from matplotlib.figure import Figure  #import Figure artist
fig=Figure()
canvas= FigureCanvas(fig)
```

```python
import numpy as np
x=np.random.randn(10000)

ax=fig.add_subplot(111)  #creates a new grid where 1,1 indicates the axes and 1 indicates the first cell of the grid which indicates the axes
```

```python
ax.hist(x,100) #generate the histogram of the 10000 numbers. #100 means to create 100 bins
```

```python
#add a title to the figure and save it
ax.set_title('Normal distribution with mu=0 ans std.dev=1')
fig.savefig('matplotlib_histogram.png')
```

# Scripting Layer

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.randn(10000)
plt.hist(x,100) #divide the data into 100 bins
plt.title(r'Normal distribution with $\mu=0, \sigma=1$')
plt.savefig('matplotlib_histogram.png')
plt.show()
```
