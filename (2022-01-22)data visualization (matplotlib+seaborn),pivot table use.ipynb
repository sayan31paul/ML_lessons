# Seaborn visualisation library

Created on Matplotlib. Hence it inherits the properties of matplotlib. Also it has some properties of itself

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x=['sun','mon','fri','sat','tue','wed','thu']
y=[5,6.7,4,6,2,4.9,1.8]
```

```python
ax=sns.barplot(x,y)                        #creating instance 'ax' from stripplot class
ax.set(xlabel='Days',ylabel='Amount_spend')  # .set() is a method in striplot class
plt.title('My first plot')                   #using matplotlib with seaborn library
plt.show()
```

```python
ax=sns.stripplot(x,y)                        #creating instance 'ax' from stripplot class
ax.set(xlabel='Days',ylabel='Amount_spend')  # .set() is a method in striplot class
plt.title('My first plot')                   #using matplotlib with seaborn library
plt.show()
```
