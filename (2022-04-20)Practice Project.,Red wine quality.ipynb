```python
import statsmodels.api as sm
from scipy import stats
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(x1):
    vif=pd.DataFrame()   #creating a vacant dataFrame
    vif['variables']=x1.columns        #add 'variables' column to the vacant DataFrame and put the column names form x1 DataFrame into it.
    vif['VIF factor']=[variance_inflation_factor(x1.values,i) for i in range(x1.shape[1])]
    return(vif)
```

# Starting to check for best models

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
lr=LogisticRegression()
parameters={'random_state':list(range(0,100))}
gd_sr=GridSearchCV(estimator=lr, param_grid= parameters, scoring='accuracy')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

```python
dtc=DecisionTreeClassifier()
parameters_dtc={'random_state':list(range(0,50)),'criterion':['gini','entropy'],'max_depth':[4,5,6,7,8,9,10]}
gd_sr=GridSearchCV(estimator=dtc, param_grid= parameters_dtc, scoring='accuracy')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

```python
rf=RandomForestClassifier()
parameters_rc={'random_state':list(range(30,50)),'n_estimators': [10,20,30],'max_depth' : [8,9,10,13,16,20],'criterion' :['gini', 'entropy']}
gd_sr=GridSearchCV(estimator=rf, param_grid= parameters_rc, scoring='accuracy')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

```python
#checking cross validation score of logistic regression with the hyperparameters obtained from GridSearchCV
lr=LogisticRegression()


from sklearn.model_selection import cross_val_score
score=cross_val_score(lr,x,y,cv=5)
print(score)
print(score.mean())
print(score.std())
```

```python
#checking cross validation score of DecisionTreeClassifier with the hyperparameters obtained from GridSearchCV
dtc=DecisionTreeClassifier(criterion='entropy', max_depth=10)

from sklearn.model_selection import cross_val_score
score=cross_val_score(dtc,x,y,cv=5)
print(score)
print(score.mean())
print(score.std())
```

```python
#checking cross validation score of RandomForestClassifier with the hyperparameters obtained from GridSearchCV
rf=RandomForestClassifier(criterion='entropy', max_depth=13, n_estimators= 30)

from sklearn.model_selection import cross_val_score
score=cross_val_score(rf,x,y,cv=5)
print(score)
print(score.mean())
print(score.std())
```

# Trying to increase the accuracy of RandomForestClassifier

```python
random_state=[]
acc_score=[]

for i in range(0,100):
    train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=i)
    rf=RandomForestClassifier(criterion= 'entropy', max_depth= 13, n_estimators=30)
    rf.fit(train_x,train_y)
    pred_y=rf.predict(test_x)
    acc=accuracy_score(test_y,pred_y)
    random_state.append(i)
    acc_score.append(acc)
```

```python
acc_score_1=pd.DataFrame()
acc_score_1['random_state']=random_state
acc_score_1['acc_score']=acc_score
```
