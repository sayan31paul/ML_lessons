```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

# 2. StaleElementReferenceException

#### Example:

```python
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By 
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get('https://www.amazon.in/')
try:
    element=driver.find_element_by_xpath("//li[@class='nav_last']")
    print(element)
except StaleElementReferenceException as e:
    print("Exception Raised:",e)
    print("Refreshing the page!")
    driver.get("https://www.amazon.in/")
    delay=10
    WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME,'nav_last')))
    element=driver.find_element_by_xpath("//li[@class='nav_last']")
    print(element)
```

```python
driver.close()
```

# 3. ElementNotVisibleException

```python
from selenium.common.exceptions import ElementNotVisibleException
import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get('https://www.icc-cricket.com/')
time.sleep(5)
try:
    element=driver.find_element_by_xpath("//li[@class='global-navigation__drop-down-item']/a")
    print(element.get_attribute('href'))
except ElementNotVisibleException as e:
    print("Exception raised:",e)
```

```python
driver.close()
```

# 4.ElementNotInteractableException

#### Handling:

```python
from selenium.common.exceptions import ElementNotInteractableException
```

```python
driver=webdriver.Chrome("chromedriver.exe")
driver.get('https://www.amazon.in/')
time.sleep(5)
try:
    element=driver.find_element_by_xpath("//i[@class='a-icon a-icon-previous-rounded']/span")
    element.click()
except ElementNotVisibleException as e:
    print("Exception raised:",e)
    element=driver.find_element_by_xpath("//a[@class='a-carousel-goto-nextapge']")
    driver.get(element.get_attribute('href'))
```

# 5.SessionNotCreatedException

```python
from selenium.common.exceptions import SessionNotCreatedException
try:
    driver=webdriver.Chrome("chromedriver.exe")
    driver.get("https://www.google.com/")
except SessionNotCreatedException as e:
    print("Exception raised:",e)
    driver=webdriver.Chrome("chromedriver.exe")
    driver.get("https://www.google.com/")
```

```python
driver.close()
```

# Selenium Web scraping example

```python
import selenium
import pandas as pd
import time

from selenium import webdriver

from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

import requests

import re

import time
```

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get('https://www.naukri.com/')
```

```python
search_field_designation=driver.find_element_by_class_name("suggestor-input ")
search_field_location=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[5]/div/div/div/input')
search_field_designation.send_keys('data Scientist')
search_field_location.send_keys('Bangalore')

search_button=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search_button.click()

time.sleep(4)
```

```python
#fetching url to open each tab
job_opening_url=[]
url=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
for i in url:
    job_opening_url.append(i.get_attribute('href'))
```

```python
len(job_opening_url)
```

```python
JD=[]

for i in job_opening_url[0:5]:
    driver.get(i)
    time.sleep(5)
    try:
        description=driver.find_element_by_xpath('/html/body/div[1]/main/div[2]/div[2]/section[2]')
        JD.append(description.text)
    except NoSuchElementException:
        JD.append('Not present')
```
