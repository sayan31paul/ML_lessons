# KMeans Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df=df[['petal length','sepal length']]
```

```python
df
```

```python
model=KMeans(n_clusters=3) #Initialising object from class.
y_model=model.fit_predict(df) #Both fitting and prediction is occuring together since it is an unsupervised ML algortihm.
y_model
```

```python
model.labels_
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=model.labels_,cmap='viridis')
```

```python
pd.value_counts(y_model)
```

```python
df[y_model==0]
```

```python
df[y_model==1]
```

# Hierarchical Agglomerative clustering

```python
from sklearn.cluster import AgglomerativeClustering

from sklearn.datasets import load_iris
iris=load_iris()

clustering=AgglomerativeClustering(n_clusters=3) #initiating object from class.
clustering.fit(iris.data) #Note that we are using only the attributes for training, not the targets.

clustering.labels_
```

```python
import matplotlib.pyplot as plt

plt.scatter(iris.data[:,2],iris.data[:,3],c=clustering.labels_,cmap='viridis')
```

```python
pd.DataFrame(iris.data)
```

# DBSCAN- Density based spatial clustering of applications with noise

```python
from sklearn.cluster import DBSCAN

from sklearn.datasets import load_iris
iris=load_iris()

db=DBSCAN().fit(iris.data) #initiating object db from class DBSCAN, also training it with unlabelled training data.

db.labels_
```

```python
plt.scatter(iris.data[:,2],iris.data[:,3],c=db.labels_,cmap='viridis')
```

```python
#DBSCAN by giving the parameters like epsilon and minimum points.
from sklearn.cluster import DBSCAN

from sklearn.datasets import load_iris
iris=load_iris()

db=DBSCAN(eps=0.5, min_samples=5)
db.fit(iris.data)

db.labels_
```
