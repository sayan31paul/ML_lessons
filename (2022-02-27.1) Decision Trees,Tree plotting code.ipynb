```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
data
```

```python
data['Gender'].replace({'Male':10,'Female':1},inplace=True)
```

```python
X=data.iloc[:,1:4].values #.values converts DataFrame to array
y=data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X=scaler.fit_transform(X) #scaler is the object name and fit_transform is the method
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
clf.score(X_train,y_train) #How well the data has fit into the model
```

```python
param_dist={'criterion':['gini','entropy'],'max_depth':[1,2,3,4,5,6,7,None]}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(clf,param_grid=param_dist,cv=10,n_jobs=-1) #cv=10 causes k fold cross validation with 10 folds
#n_jobs=-1 means use all the cores of my processor to jo this job
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_ #ACCURACY INCERASED
```
