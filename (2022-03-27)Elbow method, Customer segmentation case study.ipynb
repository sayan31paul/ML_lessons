```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df=df[['petal length','sepal length']]
```

```python
df
```

```python
wcss=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    wcss.append(km.inertia_)

plt.plot(K,wcss,marker='x')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

# Customer segmentation case study

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('MALL_customers.csv')
dataset.head(10)
```

```python
X=dataset.iloc[:,3:]
X.tail(8)
```

```python
#Elbow method

#k-means++ selects initial cluster centres for K-mean clustering in a smart way to speed up convergence

from sklearn.cluster import KMeans

wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss,marker='x')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Fitting KMeans to the dataset

kmeans=KMeans(n_clusters=5,init='k-means++') #object initialization from class
y_kmeans=kmeans.fit_predict(X)
print(y_kmeans)
```
