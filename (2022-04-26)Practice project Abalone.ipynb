# Testing different regession models

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import GridSearchCV
```

Tring to find out the best polynomial for Linear Regression

```python
poly = PolynomialFeatures(1)
X_=poly.fit_transform(x)
lr =LinearRegression()
lsscore=cross_val_score(lr,X_,y,cv=5)
print(lsscore.mean())
```

```python
poly = PolynomialFeatures(2)
X_=poly.fit_transform(x)
lr =LinearRegression()
lsscore=cross_val_score(lr,X_,y,cv=5)
print(lsscore.mean())
```

```python
poly = PolynomialFeatures(3)
X_=poly.fit_transform(x)
lr =LinearRegression()
lsscore=cross_val_score(lr,X_,y,cv=5)
print(lsscore.mean())
```

```python
poly = PolynomialFeatures(4)
X_=poly.fit_transform(x)
lr =LinearRegression()
lsscore=cross_val_score(lr,X_,y,cv=5)
print(lsscore.mean())
```

Lets try out polynomial linear regresison

```python
poly = PolynomialFeatures(2)
X_=poly.fit_transform(x)
lr=LinearRegression()
scores = cross_val_score(lr, X_, y, cv=5)
print('Cross Validation score:',scores.mean())
```

Trying Ridge regression

```python
r=Ridge()
parameters={'alpha':[0.1,0.01,0.001],'random_state':list(range(0,66))}
gd_sr=GridSearchCV(estimator=r, param_grid= parameters, scoring='r2')
gd_sr.fit(X_,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

Trying Lasso Regression

```python
l=Lasso()
parameters={'alpha':[0.1,0.01,0.001],'random_state':list(range(0,80))}
gd_sr=GridSearchCV(estimator=l, param_grid= parameters, scoring='r2')
gd_sr.fit(X_,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

Applying GridSearchCV on KNN regressor 

```python
knr=KNeighborsRegressor()
parameters={'n_neighbors':[3,5,7,9,11,15,20,25,30]}
gd_sr=GridSearchCV(estimator=knr, param_grid= parameters,scoring='r2')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

Trying RandomForestRegressor

```python
rfr=RandomForestRegressor()
parameters={'n_estimators':[30,50,70,100],'max_depth':[10,13,15,20]}
gd_sr=GridSearchCV(estimator=rfr, param_grid= parameters,scoring='r2')
gd_sr.fit(x,y)
best_parameters=gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

# Building the possible model

```python
rfr=RandomForestRegressor(n_estimators=50,max_depth=10)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=35,test_size=0.20)
rfr.fit(x_train,y_train)
y_pred=rfr.predict(x_test)
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,y_pred)))
print('r2_score:',r2_score(y_test,y_pred))
```

```python
l=Lasso(alpha=0.001)
poly = PolynomialFeatures(2)
X_=poly.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(X_,y,random_state=4,test_size=0.20)
l.fit(x_train,y_train)
y_pred=l.predict(x_test)
print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,y_pred)))
print('r2_score:',r2_score(y_test,y_pred))
```

Trying to improve the accuracy of lasso regreesion based model

```python
l=Lasso(alpha=0.001)
poly = PolynomialFeatures(2)
X_=poly.fit_transform(x)

random_state=[]
r2Score=[]

for i in list(range(0,80)):
    x_train,x_test,y_train,y_test=train_test_split(X_,y,random_state=i,test_size=0.20)
    l.fit(x_train,y_train)
    y_pred=l.predict(x_test)
    random_state.append(i)
    r2Score.append(r2_score(y_test,y_pred))
    
acc=pd.DataFrame()
acc['random_state']=random_state
acc['r2score']=r2Score
acc[(acc['r2score']==acc['r2score'].max())]
```

# Making the final model

```python
l=Lasso(alpha=0.001)
poly = PolynomialFeatures(2)
X_=poly.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(X_,y,random_state=32,test_size=0.20)
l.fit(x_train,y_train)
y_pred=l.predict(x_test)
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,y_pred)))
print('r2_score:',r2_score(y_test,y_pred))
```
